```python
#pykrx 데이터 끌고오기
import pandas as pd
import numpy as np
from pykrx import stock
from pykrx import bond
```

```python
data = pd.read_csv("../데이터자료/윈저데이터셋.csv", encoding='cp949')
data.head()
```

```python
# 거래소 코드 6개로 맞춰주기
for i in data.index:
    if len(str(data.loc[i,"거래소코드"])) == 3:
        data.loc[i,"거래소코드"]='000'+str(data.loc[i,"거래소코드"])
    elif len(str(data.loc[i,"거래소코드"])) == 4:
        data.loc[i,"거래소코드"]='00'+str(data.loc[i,"거래소코드"])
    elif len(str(data.loc[i,"거래소코드"])) == 5:
        data.loc[i,"거래소코드"]='0'+str(data.loc[i,"거래소코드"])
```

```python
#거래소 코드 개별로 추출
a =data["거래소코드"].unique()
a = pd.DataFrame(a)
a.columns=["거래소코드"]
len(a)
```

```python
#회사명 추출
c = data["회사명"].unique()
c = pd.DataFrame(c)
c.columns=["회사명"]
len(c)
```

```python
#데이터프레임 일치화
a["회사명"] = c["회사명"]
```

```python
# 코드와 회사명
a
```

```python
# 코스닥 개별기업들의 일자별 주가 추출
b=[]
for i in a.index:
    df = stock.get_market_ohlcv("20110101", "20211231", a.loc[i,"거래소코드"])
    b.append(df)
```

```python
# 각 인덱스에 회사명이랑 거래소코드 넣어주기
for i in a.index:
    b[i]["거래소코드"]=a.loc[i,"거래소코드"]
    b[i]["회사명"] =a.loc[i,"회사명"]
```

```python
# 기업별 데이터프레임 합하기
data_all = pd.concat(b)
```

```python
data_all=data_all.reset_index()
data_all
```
