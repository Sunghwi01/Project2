```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./dataset/label지정.csv", encoding="cp949")
```

```python
df2 = df.copy()
# df2 = df2[df2['label'] == 0]
df2 = df2[['[제조]유동자산구성비율', '[제조]순운전자본비율','[제조]총자본증가율', '[제조]매출채권회전률', '[제조]자본분배율', '[제조]매입채무회전기간', '[제조]비유동자산증가율', '[제조]매출액순이익률','[제조]자기자본순이익률','year','label']]
df2
```

```python
from scipy.stats.mstats import winsorize
import warnings
warnings.filterwarnings('ignore')

for col in df2.columns: 
     df2[col] = winsorize(df2[col], limits=0.01) # 범위. 양끝 0.01을 안쪽으로 넣어준다!)

df2.describe()
```

```python
df_n = df2[df2['year'] < 2019]
df_i = df2[df2['year'] >= 2019]
```

```python
from sklearn.model_selection import train_test_split
X = df_n.iloc[:,:-1]
y = df_n.iloc[:,-1]
# X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

```python
from imblearn.under_sampling import *

X_samp, y_samp = SMOTE(random_state=0,sampling_strategy=0.25).fit_resample(X, y)

print('RS 적용 전 학습용 피처/레이블 데이터 세트: ', X.shape, y.shape)
print('RS 적용 후 학습용 피처/레이블 데이터 세트: ', X_samp.shape, y_samp.shape)
print('RS 적용 후 레이블 값 분포: \n', pd.Series(y_samp).value_counts())
```

```python
Same_sampling_smote = pd.concat([X_samp,y_samp], axis=1)
```

```python
Same_sampling_smote.to_csv('./dataset/Same_sampling_smote.csv' , encoding='CP949', index=False)
```

```python
from imblearn.under_sampling import EditedNearestNeighbours

enn = EditedNearestNeighbours(n_neighbors = 651)

X_res, y_res = enn.fit_resample(X,y)

print('ENN적용 후 :', X_res.shape, y_res.shape)
print('ENN 적용 후 레이블 값 분포 : \n', pd.Series(y_res).value_counts())
```

```python
Same_sampling_ENN = pd.concat([X_res,y_res], axis=1)
```

```python
Same_sampling_ENN.to_csv('./dataset/Same_sampling_ENN.csv', encoding='CP949', index=False)
```

```python
# imbalanced-learn 패키지
from imblearn.combine import SMOTEENN

# SMOTE 객체 생성
smt = SMOTEENN(random_state=0,sampling_strategy=0.2483)

X_smtn,y_smtn = smt.fit_resample(X,y)

# 샘플링
print('SMOTEENN 적용 전 학습용 피처/레이블 데이터 세트: ', X.shape, y.shape)
print('SMOTEENN 적용 후 학습용 피처/레이블 데이터 세트: ', X_smtn.shape, y_smtn.shape)
print('SMOTEENN 적용 후 레이블 값 분포: \n', pd.Series(y_smtn).value_counts())
```

```python
Same_sampling_smotEEN = pd.concat([X_smtn,y_smtn], axis=1)
```

```python
Same_sampling_smotEEN.to_csv('./dataset/Same_sampling_smotEEN.csv', encoding='CP949', index=False)
```
