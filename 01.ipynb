```python
import pandas as pd
```

```python
kospi = pd.read_csv('./KOSPI/KOSPI.csv', encoding='CP949')
```

```python

kospi['year'] = pd.to_datetime(kospi['회계년도']).dt.year
kospi['month'] = pd.to_datetime(kospi['회계년도']).dt.month
```

```python
mdx = kospi[kospi['month'] != 12].index
kospi = kospi.drop(mdx)
```

```python

ldx = kospi[kospi['산업명'] == '  기타금융업'].index
ldx1 = kospi[kospi['산업명'] == '  보험업'].index
ldx2 = kospi[kospi['산업명'] == '  신용금고업'].index
ldx3 = kospi[kospi['산업명'] == '  여신금융업'].index
ldx4 = kospi[kospi['산업명'] == '  은행업'].index
ldx5 = kospi[kospi['산업명'] == '  종합금융업'].index
ldx6 = kospi[kospi['산업명'] == '  증권업'].index

kospi = kospi.drop(ldx)
kospi = kospi.drop(ldx1)
kospi = kospi.drop(ldx2)
kospi = kospi.drop(ldx3)
kospi = kospi.drop(ldx4)
kospi = kospi.drop(ldx5)
kospi = kospi.drop(ldx6)
```

```python
kospi['month'].unique()
```

```python
kospi['연속매출액증가율'] = ''
kospi['연속ROA'] = ''
kospi['연속ROE'] = ''
kospi['1년연속회사인지'] = kospi['회사명'].shift(1)
kospi['2년연속회사인지'] = kospi['회사명'].shift(2)
kospi['1년전매출액증가율'] = kospi['매출액증가율(IFRS연결)'].shift(1)
kospi['2년전매출액증가율'] = kospi['매출액증가율(IFRS연결)'].shift(2)
kospi['1년전ROA'] = kospi['총자본순이익률(IFRS연결)'].shift(1)
kospi['2년전ROA'] = kospi['총자본순이익률(IFRS연결)'].shift(2)
kospi['1년전ROE'] = kospi['자기자본순이익률(IFRS연결)'].shift(1)
kospi['2년전ROE'] = kospi['자기자본순이익률(IFRS연결)'].shift(2)
```

```python
kospi.reset_index(drop=True,inplace=True)
```

```python
kospi
```

```python
for i in range(len(kospi)):
    if kospi['회사명'][i] == kospi['2년연속회사인지'][i]:
        if kospi['매출액증가율(IFRS연결)'][i] < 0 and kospi['1년전매출액증가율'][i] < 0 and kospi['2년전매출액증가율'][i] < 0:
            kospi['연속매출액증가율'][i] = '3년연속매출액증가마이너스'
        else :
            kospi['연속매출액증가율'][i] = ''
    else :
        kospi['연속매출액증가율'][i] = ''
```

```python
for i in range(len(kospi)):
    if kospi['회사명'][i] == kospi['2년연속회사인지'][i]:
        if kospi['자기자본순이익률(IFRS연결)'][i] < 0 and kospi['1년전ROE'][i] < 0 and kospi['2년전ROE'][i] < 0:
            kospi['연속ROE'][i] = '3년연속ROE마이너스'
        else :
            kospi['연속ROE'][i] = ''
    else :
        kospi['연속ROE'][i] = ''
```

```python
for i in range(len(kospi)):
    if kospi['회사명'][i] == kospi['2년연속회사인지'][i]:
        if kospi['총자본순이익률(IFRS연결)'][i] < 0 and kospi['1년전ROA'][i] < 0 and kospi['2년전ROA'][i] < 0:
            kospi['연속ROA'][i] = '3년연속ROA마이너스'
        else :
            kospi['연속ROA'][i] = ''
    else :
        kospi['연속ROA'][i] = ''
```
