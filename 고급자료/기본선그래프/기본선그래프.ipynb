```python
import pandas as pd

a=pd.read_excel('우리나라GDP.xls')
b=a.T
header = b.iloc[0]
b.rename(columns=header, inplace=True)
b.drop(b.index[0],axis=0,inplace=True)
b.reset_index(inplace=True)
b.rename(columns={"index":"날짜"},inplace=True)
b["국내총생산(명목GDP)"]=b["국내총생산(명목GDP)"].str.replace(",","")
b=b.astype(float)
b=b.astype({"날짜":"int"})
b
```

```python
import matplotlib.pyplot as plt 
```

### color 색깔 변경

- color인자에 헥스코드를 넣어서 변경할 수 있고, 밑에 green처럼 색갈명을 넣어서 변경할 수도 있다.


```python
from IPython.display import Image
Image('컬러코드.png')
```

```python
# 선 그래프를 그린다. x축에는 날짜값, y축에는 "국내총생산(명목GDP)"이 표시된다.
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='green', marker='o') 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='blue', marker='o')
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='pink', marker='o')
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='#AA00FF', marker='o') 

plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='#AA00FF', marker='o') 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='k', marker='o')
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='b', marker='o')

```

### 점 모양 변경

- marker에는 o를 넣으면 점표시, s를 넣으면 사각형의 점, *은 별표시의 점이 들어가 표현된다. 

```python
# 선 그래프를 그린다. x축에는 날짜값, y축에는 "국내총생산(명목GDP)"이 표시된다.
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='pink', marker='o') 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='pink', marker='s') 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color='pink', marker='*') 
plt.show
```

### linestyle을 통해서 선모양을 바꿔줄 수 있음

```python
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='dashed')
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='--')

plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='dashdot')
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='-.')

plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='solid') # solid가 기본
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='-') 

plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle='dotted') 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], linestyle=':') 

plt.show
```

#### color와 linestyle을 따로 안 넣어주고 밑에와 같이 표현할 수 있음

```python
plt.plot(b["날짜"], b["국내총생산(명목GDP)"],"k--") 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"],"b:") 
plt.plot(b["날짜"], b["국내총생산(명목GDP)"],"r-") 

plt.show
```

### subplot과 subplots의 차이점
 - 두개 모두 그래프를 한번에 여러개 보여주기 위해 사용함
 - subplot는 일일이 다 설정해줘야하지만

 - subplots는 한번에 설정해 줄 수 있다.

### subplot에서는 괄호안에 위치를 즉 좌표를 하나씩 구현한다고 생각을 하면 됨

```python
Image('subplot용.jpg')
```

```python
# subplot의 예
plt.subplot(2,1,1)
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color = 'red', alpha = 0.5)

plt.subplot(2,1,2)
plt.plot(b["날짜"],  b["경제성장률(실질GDP성장률)"], color = 'blue', alpha = 0.5)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(b["날짜"], b["국내총생산(명목GDP)"], color = 'red')

plt.subplot(1, 2, 2)
plt.plot(b["날짜"], b["경제성장률(실질GDP성장률)"], color = 'pink')
```

### subplots에서는 괄호안에 행,렬 수만 정해주면 됨

```python
# subplots의 예
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(b["날짜"], b["국내총생산(명목GDP)"], color = 'red')
axes[1].plot(b["날짜"], b["경제성장률(실질GDP성장률)"], color = 'pink')

plt.show()
```

```python
# subplots의 예
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(b["날짜"], b["국내총생산(명목GDP)"], color = 'red')
axes[1].plot(b["날짜"], b["경제성장률(실질GDP성장률)"], color = 'pink')

plt.show()
```
