```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

flights= pd.read_csv("flights.csv",index_col=0)
flights

```

```python
plt.pcolor(flights)
plt.xticks(np.arange(len(flights.columns)),flights.columns)
plt.yticks(np.arange(len(flights.index)), flights.index)
plt.title('Heatmap by plt.pcolor()', fontsize=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.colorbar()
```

# seaborn으로 히트맵 그리기

```python
import seaborn as sns
df=flights
ax = sns.heatmap(df,annot=True,fmt='d',cmap='Blues') #annot를 통해서 숫자를 넣을 수 있음
                                        # fmt='d' 는 정수 형태(integer format)로 숫자를 입력하라는 뜻입니다. 
                                        #  fmt = '.2f' 형태로 입력하게 되면 소수점 둘째자리까지 나오게 할 수 있음

plt.title('Heatmap of Flight by seaborn', fontsize=20)
plt.show() 
```
