```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris2 = iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].mean()

plt.figure(figsize=(20,5))
iris2['sepal_length'].plot.bar(x=iris2.index, color='green', alpha=0.5)
plt.title('Average Sepal Length by Species', fontsize=15)
plt.xlabel('species', fontsize=13)
plt.ylabel('average sepal length', fontsize=13)
plt.xticks(fontsize=12, rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris2 = iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].mean()
iris4 = iris2.stack().reset_index()
iris4.columns = ['species','class','size']

plt.figure(figsize=(20,5))
sns.barplot(x='species', y='size', hue='class', data=iris4, palette='Reds', alpha=0.8)
plt.show()
```

### 1번 문제 컬럼 바꾸기

```python

tips.columns=["총액","팁","성별","흡연여부","요일","시간대","인원"]
tips
```

```python
tips["성별"].unique()
```

```python
tips["성별"]=tips["성별"].replace(["Female","Male"],["여성","남성"])
```

```python
tips["요일"].unique()
```

```python
tips["요일"]=tips["요일"].replace(['Sun', 'Sat', 'Thur', 'Fri'],['일요일', '토요일', '목요일', '금요일'])
```

```python
tips["시간대"].unique()
```

```python
tips["시간대"]=tips["시간대"].replace(['Dinner', 'Lunch'],['저녁', '점심'])
```

```python
tips
```

```python
tips.value_counts("성별")
```
