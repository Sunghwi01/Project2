```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris2 = iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].mean()

plt.figure(figsize=(20,5))
iris2['sepal_length'].plot.bar(x=iris2.index, color='green', alpha=0.5)
plt.title('Average Sepal Length by Species', fontsize=15)
plt.xlabel('species', fontsize=13)
plt.ylabel('average sepal length', fontsize=13)
plt.xticks(fontsize=12, rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris2 = iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].mean()

plt.figure(figsize=(20,5))
iris2['sepal_length'].plot.barh(x=iris2.index, color='green', alpha=0.5)
plt.title('Average Sepal Length by Species', fontsize=15)
plt.xlabel('average sepal length', fontsize=12)
plt.ylabel('species', fontsize=12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris2 = iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].mean()
iris3 = iris2.reset_index()

iris3.plot.bar(x='species', alpha=0.5, figsize=(20,5))
plt.title('Average Sepal Length by Species', fontsize=15)
plt.xlabel('species', fontsize=13)
plt.ylabel('average sepal length', fontsize=13)
plt.xticks(fontsize=12, rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris2 = iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].mean()
iris4 = iris2.stack().reset_index()
iris4.columns = ['species','class','size']

plt.figure(figsize=(20,5))
sns.barplot(x='species', y='size', hue='class', data=iris4, palette='Reds', alpha=0.8)
plt.show()
```
