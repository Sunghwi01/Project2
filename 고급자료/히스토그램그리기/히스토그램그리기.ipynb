```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
        80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]
```

# histtype을 통해 종류별로 그릴 수 있음

```python
plt.hist((weight, weight2), histtype='bar',bins=10,density=True, alpha=0.3)  # 소괄호를 넣어주는 것 이 특징임 , 알파는 투명도 bin은 몇구간으로 쪼개갰는가?
plt.title('histtype - bar')
plt.figure()

plt.hist((weight, weight2), histtype='barstacked')
plt.title('histtype - barstacked')
plt.figure()

plt.hist((weight, weight2), histtype='step')
plt.title('histtype - step')
plt.figure()

plt.hist((weight, weight2), histtype='stepfilled')
plt.title('histtype - stepfilled')
plt.show()
```

# seaborn으로 그래프그리기

```python
import seaborn as sns

ax = sns.distplot(weight,hist=True,kde=False, bins=10, color='blue',hist_kws={'edgecolor': 'gray'}, 
                  # kde는 여기에서 커널밀도함수를 그리라는 뜻
                  # kde True는 
                  kde_kws={'linewidth': 2})
ax.set_title('무게별 차지하는')
ax.set_xlabel('무게')
ax.set_ylabel('Density')
plt.show()

plt.show()
```

# 실습

```python
iris=pd.read_csv("iris_data.csv",index_col=0)
iris
```

```python
iris.groupby('species').size()
```

```python
plt.hist(iris['sepal_width'], bins=15)
# plt.hist(iris['sepal_width'], bins=10, density=True)  ### y축을 빈도수가 아니라 density로 하고 싶을 때는 density =True를 설정해주면 됩니다.
```

```python
plt.hist(iris['sepal_width'], bins=10,density=True, facecolor='blue', alpha=0.3) # 여기에서는 color가 아니라 facecolor하고 설정하고, alpha값은 투명도설정

plt.xlabel('X bins')
plt.ylabel('Density')
plt.title('Histogram of Sepal Width')
plt.axis([1.5, 4.5, 0, 1.1])
plt.show()
```
