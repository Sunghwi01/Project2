### 전체기업 이름에서 (주) 제거

- (주)의 주 뿐만 아니라 그냥 기업이름의 주도 빠진다고 했으니 사용 자제하는걸로

```python
kosdaq['회사명']= kosdaq['회사명'].str.replace('\(','')
kosdaq['회사명']= kosdaq['회사명'].str.replace('\)','')
kosdaq['회사명']= kosdaq['회사명'].str.replace('주','')
```

## 소속코드가 4인 기업들만 추려내서 몇개 나오는지 확인

### 소속코드가 4이면 상장폐지기업, 5이면 코스닥기업, 6이면 관리종목

```python
ldx = kosdaq[kosdaq['소속코드'] != 4].index
kosdaq1 = kosdaq.drop(ldx)
```

```python
kosdaq['회계년도'] = pd.to_datetime(kosdaq['회계년도'])
kosdaq['상장폐지일'] = pd.to_datetime(kosdaq['상장폐지일'])
kosdaq['상장일'] = pd.to_datetime(kosdaq['상장일'])
```

```python
kosdaq['회계년'] = kosdaq['회계년도'].dt.year
kosdaq['회계월'] = kosdaq['회계년도'].dt.month
kosdaq['상폐년도'] = kosdaq['상장폐지일'].dt.year
kosdaq['상장년'] = kosdaq['상장일'].dt.year
```
