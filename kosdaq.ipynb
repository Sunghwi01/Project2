# 전처리 과정

## 금융업 제거

```python
## 금융업 제거

ldx = kosdaq[kosdaq['산업명'] == '    기타금융업'].index
ldx1 = kosdaq[kosdaq['산업명'] == '  보험업'].index
ldx2 = kosdaq[kosdaq['산업명'] == '  신용금고업'].index
ldx3 = kosdaq[kosdaq['산업명'] == '    여신금융업'].index
ldx4 = kosdaq[kosdaq['산업명'] == '  은행업'].index
ldx5 = kosdaq[kosdaq['산업명'] == '  종합금융업'].index
ldx6 = kosdaq[kosdaq['산업명'] == '  증권업'].index

kosdaq = kosdaq.drop(ldx)
kosdaq = kosdaq.drop(ldx1)
kosdaq = kosdaq.drop(ldx2)
kosdaq = kosdaq.drop(ldx3)
kosdaq = kosdaq.drop(ldx4)
kosdaq = kosdaq.drop(ldx5)
kosdaq = kosdaq.drop(ldx6)
```

## 회계년월중 회계월이 12월인 기업만 선정

```python
## 회계월이 12월을 제외한 다른 달일 경우 배제

kosdaq['year'] = pd.to_datetime(kosdaq['회계년도']).dt.year
kosdaq['month'] = pd.to_datetime(kosdaq['회계년도']).dt.month
mdx = kosdaq[kosdaq['month'] != 12].index
kosdaq = kosdaq.drop(mdx)
```

```python
kosdaq['회사명']= kosdaq['회사명'].str.replace('\(','')
kosdaq['회사명']= kosdaq['회사명'].str.replace('\)','')
kosdaq['회사명']= kosdaq['회사명'].str.replace('주','')
```
